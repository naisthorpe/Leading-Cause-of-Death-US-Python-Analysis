```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
lcd = pd.read_csv('NCHS_-_Leading_Causes_of_Death__United_States.csv')
```

### Viewing my data to see which columns to keep and remove

###### I don't need the 113 Cause Name data, and the Age-adjusted Death Rate is above my head as to what it actually means

```python
lcd.head()
```

#### Dropping 113 Cause Name & Age-adjusted Death Rate columns

```python
lcd.drop(columns='113 Cause Name', inplace=True)
```

```python
lcd.head()
```

```python
lcd.drop(columns='Age-adjusted Death Rate', inplace=True)
```

```python
lcd.head()
```

### Now I'd like to look only at North Carolina

```python
lcd[lcd['State'] == "North Carolina"].head().sort_values(by=['Deaths'], ascending=False)
```

#### Okay, let's drop All Causes for a second

```python
lcd.drop(lcd.loc[lcd['Cause Name'] == 'All causes'].index, inplace=True)
```

```python
lcd.head()
```

### Setting an NC dataframe 

```python
lcd_nc = lcd[lcd["State"] == 'North Carolina']
```

```python
lcd_nc.head()
```

### Testing out sorting by value 

```python
lcd_nc_99 = lcd_nc[lcd["Year"] == 1999].sort_values(by=['Deaths'], ascending=False)
```

```python
lcd_nc_99
```

### Here I am sorting by Year and Deaths (both descending), then grouping by Year and displaying the top result

```python
lcd.sort_values(by=["Year", "Deaths"], ascending=False).groupby("Year").head(1)        
```

### Setting an NC mask 

```python
nc_mask = lcd["State"] == "North Carolina"   
```

### Let's look at the top Cause of Death per year in NC

```python
lcd[nc_mask].sort_values(by=["Year", "Deaths"], ascending=False).groupby("Year").head(1)
```

```python
lcd_states_only = lcd[lcd["State"] != "United States"]
```

```python
lcd_states_only.head()
```

```python
lcd_states_only[nc_mask].sort_values(by=["Year", "Deaths"], ascending=False).groupby("Year").head(1)
```

# Now I'd like to look at trends for specific causes of death

### I'll start with looking at Cancer  

```python
cancer_mask = lcd["Cause Name"] == 'Cancer'
```

```python
lcd_states_only[nc_mask][cancer_mask].sort_values(by=['Year', 'Deaths'], ascending=False)
```

```python
lcd_nc_cancer = lcd_states_only[nc_mask][cancer_mask].sort_values(by=['Year', 'Deaths'], ascending=False)
```

```python
lcd_nc_cancer
```

#### Time for a Cancer in NC bar graph

```python
plt.bar(lcd_nc_cancer["Year"], lcd_nc_cancer["Deaths"], width=0.5, color="red")
plt.show()
```

### Now I'd like to look at Alzheimer's numbers

```python
alz_mask = lcd["Cause Name"] == "Alzheimer's disease"
```

```python
lcd_nc_alz = lcd_states_only[nc_mask][alz_mask].sort_values(by=['Year', 'Deaths'], ascending=False)
```

```python
lcd_nc_alz
```

#### Alzheimer's in NC Graph

```python
plt.bar(lcd_nc_alz["Year"], lcd_nc_alz["Deaths"], width=0.5, color="red")
plt.show()
```

## So what's the average number of cancer deaths from 1999 - 2017 in NC?

```python
lcd_nc_cancer["Deaths"].mean()
```
