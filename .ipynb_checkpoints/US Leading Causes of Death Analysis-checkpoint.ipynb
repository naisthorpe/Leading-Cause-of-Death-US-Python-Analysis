# Now I'd like to rank each cause of death per year in NC from 1-10 based on number of deaths

```python
nc_mask = lcd["State"] == 'North Carolina'
```

```python
lcd.drop(lcd.loc[lcd['Cause Name'] == 'All causes'].index, inplace=True)
```

```python
lcd_nc = lcd[nc_mask]
```

```python
lcd_nc.sort_values(by=['Year', 'Deaths'], ascending=False, inplace=True)
```

```python
lcd_nc.groupby("Year")
```

```python
lcd_nc['Rank'] = lcd_nc.groupby("Year")["Deaths"].rank("first", ascending=False)
```

```python
year16_mask = lcd["Year"] == 2016
```

```python
lcd_nc[year16_mask]
```

```python
# looking at the top 3 per year
lcd_nc.groupby("Year").head(3)
```

```python
def float_to_str(rank):
    intrank = int(rank)
    year = lcd["Year"].copy()
    return "# " + str(intrank) + ' in '
```

```python
lcd_nc["Rank"] = lcd_nc['Rank'].apply(float_to_str)
```

```python
lcd_nc
```

```python
# setting up a copy of Year as a string instead of integer
add_year = lcd['Year'].apply(str).copy()
```

```python
# adding the year to the end of the Rank string
lcd_nc['Rank'] = lcd_nc['Rank'].str.cat(add_year, sep = '')
```

```python
lcd_nc
```
